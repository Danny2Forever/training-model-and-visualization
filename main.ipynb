````python
# นำเข้าเครื่องมือที่จำเป็น
import numpy as np
from sklearn import datasets

#สร้าง datasets
np.random.seed(10)

# x เก็บพิกัดข้อมูล 1000 จุด
# y เก็บกระจุกข้อมูลจาก make_blobs

x,y = datasets.make_blobs( n_samples=1000 , n_features=2 , centers=5 , cluster_std=2)
````

````python
# ตรวจสอบข้อมูลในตัวแปร x
print(x)
````

````python
# ตรวจสอบข้อมูลในตัวแปร y
print(y)
````

````python
import matplotlib.pyplot as plt

plt.gca().scatter(x[:,0],x[:,1] , c=y , cmap ='winter' , edgecolor='k')
plt.show()
````

**เทรนโมเดล**
> K-Means เป็นอัลกอริทึม

````python
from sklearn.cluster import KMeans
# นำมาสร้างเป็นตัวแปร และกำหนด output
km = KMeans(n_clusters=10)
km.fit(x)
````

````python
z = km.predict(x)
````

````python
print(z)
````

**ฟล็อตกราฟ 2 มิติ**

````python
import matplotlib.pyplot as plt

plt.gca().scatter(x[:,0],x[:,1] , c=z , edgecolor = 'k' ,cmap = 'jet')
plt.show()
````

````python
import matplotlib.pyplot as plt

plt.gca().scatter(x[:,0],x[:,1] , c=z , edgecolor = 'k' ,cmap = 'jet')
plt.scatter(km.cluster_centers_[:,0] , km.cluster_centers_[:,1] , 100 , 'red')
plt.show()
````

**สร้างกราฟ 3 มิติ**

````python
x2 = km.transform(x)
print(x2)
````

````python
cen = km.transform(km.cluster_centers_)
print(cen)
````

````python
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize = [6,6])
ax = plt.axes([0,0,1,1], projection='3d')
ax.scatter( x2[:,0] , x2[:,1] ,x2[:,2] , c=z , edgecolor='lime' , cmap = "winter")
ax.scatter( cen[:,0], cen[:,1] , cen[:,2] , s=300, c='#FE0000')
plt.show()
````
